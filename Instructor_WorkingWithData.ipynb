## Subsetting data by conditions = filtering

```python
# use conditional statements for returning rows from dataframe
# let's return all observations from village God
safi[(safi.village == "God")]
```

```python
# combine conditions (filters)
safi[(safi.village == "Ruaca") & (safi.no_membrs > 15)]
```

```python
# combine filtering and select specific columns
safi[(safi.village == "Ruaca") & (safi.no_membrs > 15)][["no_membrs","years_liv","rooms"]]
```
